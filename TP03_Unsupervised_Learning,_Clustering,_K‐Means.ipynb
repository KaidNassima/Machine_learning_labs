EXERCICE 01

QUESTION 01

```python
from sklearn.datasets import make_blobs 
import numpy as np 
import matplotlib.pyplot as plt

#setting the centers of the blobs 
blob_centers = np.array(
  [[0.2, 2.3],
  [-1.5,2.3],
  [-2.8, 1.8],
  [-2.8, 2.8],
  [-2.8, 1.3]
  ])
blob_std = np.array([0.4, 0.3, 0.1, 0.1, 0.1])
X, y = make_blobs(n_samples=2000, centers=blob_centers,cluster_std=blob_std, random_state=7)
```

QUESTION 02

```python
#plotting the blobs 
plt.scatter(X[:,0],X[:, 1], c=y)
plt.title("blobs")
plt.xlabel("features")
plt.ylabel("y")
plt.show()
#according to the figure we have 5 clusters 
```

QUESTION 03

```python
from sklearn.cluster import KMeans
#k-means clustering model :
kmeans = KMeans(n_clusters=5, random_state=0).fit(X)
```

QUESTION 04

```python
#displaying the centers of each cluster
centers = kmeans.cluster_centers_
print(centers[:,0])
```

QUESTION 05 : Visualizing the different clusters 

```python
#centers = kmeans.cluster_centers_
plt.scatter(X[:,0],X[:, 1], c=kmeans.predict(X))
plt.scatter(centers[:,0], centers[:,1], c='black')
plt.title("visualizing the clusters")
plt.xlabel("Features")
plt.ylabel("Target")

plt.show()
```

QUESTION 06

```python
#Calculating the silhouette score
from sklearn.metrics import silhouette_samples, silhouette_score
pred = kmeans.fit_predict(X)
centers = kmeans.cluster_centers_

score = silhouette_score(X, pred)
print("silhouette score is {}".format(score))
```

QUESTION 07


```python
#prediciting new instances 
X_new = np.array([[0, 2], [3, 2], [-3, 3], [-3, 2.5]])
predicted_labels = kmeans.predict(X_new)
print(predicted_labels)
```

EXERCICE 02

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans 
```

QUESTION 1 : Load Data 

```python
iris = load_iris()
X = iris.data
y = iris.target 
```

QUESTION 02 : Visualize Data

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

plt.tight_layout()
plt.show()
```

QUESTION 03 : K-MEANS CLUSTRING

```python
kmeans_2 = KMeans(n_clusters=2, random_state=0).fit(X)
centers_2 = kmeans_2.cluster_centers_

kmeans_3 = KMeans(n_clusters=3, random_state=0).fit(X)
centers_3 = kmeans_3.cluster_centers_

kmeans_4 = KMeans(n_clusters=4, random_state=0).fit(X)
centers_4 = kmeans_4.cluster_centers_

kmeans_5 = KMeans(n_clusters=5, random_state=0).fit(X)
centers_5 = kmeans_5.cluster_centers_
```

QUESTION 4 : Plot the result and calculate the silhouette score.

```python
#Plot Results
plt.scatter(X[:,0],X[:, 1], c=kmeans_2.predict(X))
plt.scatter(centers_2[:,0], centers_2[:,1], c='red')
plt.title("K = 2")
plt.xlabel("Features")
plt.ylabel("Target")
plt.show()

plt.scatter(X[:,0],X[:, 1], c=kmeans_3.predict(X))
plt.scatter(centers_3[:,0], centers_3[:,1], c='red')
plt.title("K = 3")
plt.xlabel("Features")
plt.ylabel("Target")
plt.show()

plt.scatter(X[:,0],X[:, 1], c=kmeans_4.predict(X))
plt.scatter(centers_4[:,0], centers_4[:,1], c='red')
plt.title("K = 4")
plt.xlabel("Features")
plt.ylabel("Target")
plt.show()

plt.scatter(X[:,0],X[:, 1], c=kmeans_5.predict(X))
plt.scatter(centers_5[:,0], centers_5[:,1], c='red')
plt.title("K = 5")
plt.xlabel("Features")
plt.ylabel("Target")
plt.show()
```

```python
#silhouette_score
from sklearn.metrics import silhouette_score

score_k_2 = silhouette_score(X, kmeans_2.labels_, metric='euclidean')
score_k_3 = silhouette_score(X, kmeans_3.labels_, metric='euclidean')
score_k_4 = silhouette_score(X, kmeans_4.labels_, metric='euclidean')
score_k_5 = silhouette_score(X, kmeans_5.labels_, metric='euclidean')

print("Score pour K=2 : ", score_k_2)
print("Score pour K=3 : ", score_k_3)
print("Score pour K=4 : ", score_k_4)
print("Score pour K=5 : ", score_k_5)
```

QUESTION 05 : Silhouette Score Vs K values

```python
Score = [score_k_2, score_k_3, score_k_4, score_k_5]
K = [2, 3, 4, 5]
plt.plot(K,Score)
plt.xlabel('Values of K') 
plt.ylabel('Score Silouhette') 
plt.title('Silhouette Score Vs K values')
plt.show()
```

QUESTION 6 : silhouette diagram

```python
from yellowbrick.cluster import silhouette_visualizer

silhouette_visualizer(kmeans_2, X, colors='yellowbrick')
silhouette_visualizer(kmeans_3, X, colors='yellowbrick')
silhouette_visualizer(kmeans_4, X, colors='yellowbrick')
silhouette_visualizer(kmeans_5, X, colors='yellowbrick')

#The adequate Value is K = 3 : densit rapproch + dpasse la moyenne 
```
